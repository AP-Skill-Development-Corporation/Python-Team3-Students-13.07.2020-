```python
#Area plot
from matplotlib import pyplot as plt
days=[1,2,3,4]
sleeping=[10,20,30,40]
eating=[50,60,70,20]
studying=[10,5,6,12]
playing=[15,25,35,45]
plt.plot([],[],color='r',label="sleeping",linewidth=5)
plt.plot([],[],color='b',label="eating",linewidth=5)
plt.plot([],[],color='m',label="studying",linewidth=5)
plt.plot([],[],color='c',label="playing",linewidth=5)
plt.stackplot(days,sleeping,eating,studying,playing,colors=['r','b','m','c'])
plt.legend()
plt.title("Area-plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
print(plt.show())
```

```python
##Piechart
plt.pie([10,20,30,40])
plt.show()
```

```python
#applying styles
days=[1,2,3,4,5]
sleeping=[10,20,30,40,50]
eating=[50,60,70,20,11]
studying=[10,5,6,12,21]
playing=[15,25,35,45,13]
slices=[7,2,2,13]
a1=['sleeping','eating','studying','playing']
cols=['b','g','k','y']
plt.pie(slices,labels=a1,colors=cols,startangle=90,shadow=True,autopct="%1.2f%%")
plt.legend()
plt.show()
```

#### **Class**

* It can be defined as collection of objects.It is a logical entity that hass some specific attributes and methods.
* **Syntax**:
    class class_name():
        list of vars
        list of methods
     * Rule for classname
         * class name starts with uppercase
* Ex:class Employee():Employee is the class name,then it should contain an attributes i.e name,gmail,id,etc...
* Collection of variables and methods
* function:collection of statements
* Method: if function is existed inside the class then it called as method
* class is a keyword for creating class structure
* class and objects are interrelated with each other

```python
class Math():
    #Inside math class defining add(),sub(),mul(),mod()
    def add(a,b):
        return a+b
    def sub(a,b):
        return a-b
    def mul(a,b):
        return a*b
    def mod(a,b):
        return a%b
```

```python
#object creation
cal=Math
```

```python
print(cal)
```

```python
print(dir(cal))
```

```python
print("addition",cal.add(19,7))
print("subtraction",cal.sub(33,8))
print("multiplication",cal.mul(19,8))
print("moddiv",cal.mod(35,4))
```
