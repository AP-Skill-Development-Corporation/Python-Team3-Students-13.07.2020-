### Data visualization using matplotlib
* It is plotting library for the python programming language and its numerical mathematical extension Numpy
* It is used to 2D graphics in python programming language
* It can be used in python script,shell,webapplication server
* Different toolkit:
    - ***BaseMap***
    - ***cartopy***
    - ***Excel Tools***
    - ***Mplot3d***
    - ***Natgrid***
 * Types of Graphs:
     * Line graph
     * Bar graph
     * Histogram
     * Scatter graph
     * Area plot
     * Piechart

```python
from matplotlib import pyplot as plt
```

```python
#line graph
plt.plot([1,3,8,9],[1,6,4,2])
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]
plt.plot(x,y)
plt.title("line-graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Applying styles(line graph)
from matplotlib import style
x=[5,2,7]
y=[2,16,4]
a=[6,9,11]
b=[21,54,3]
#style.use("ggplot")
plt.plot(x,y,label="A",linewidth=5)
plt.plot(a,b,label="B",linewidth=5)
plt.title("line-graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(color="k")
plt.show()
```

```python
import numpy as np
x=np.linspace(0,10)
y=np.exp(-x)
plt.plot(x,y)
plt.savefig("linegraph.jpg")
plt.show()
```

```python
#bar-graph
plt.bar([10,20,30,40],['c','c++','java','Python'])
plt.show()
```

```python
plt.bar([10,20,30,40],['c','c++','java','Python'])
plt.title("bar-graph")
plt.xlabel("Values")
plt.ylabel("Languages")
plt.show()
```

```python
plt.bar([10,30,50],[20,30,40],label="values",width=10)
plt.bar([70,90,50],[70,80,90],label="languages",width=15)
plt.title("bar-graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
plt.scatter([1,3,8,9],[1,6,4,2])
```

```python
a=[1,1.5,2,2.5,3,3.5,4,4.5]
b=[10,15,20,25,30,35,40,50]
a1=[6,6.5,7,7.5,8,8.5,9,9.5]
b1=[10,20,30,40,60,50,64,17]
plt.scatter(a,b,label="highincome low saving",color='r')
plt.scatter(a1,b1,label="low saving high income",color='b')
plt.title("scatter-graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
from numpy.random import rand
x=rand(100)
y=rand(100)
plt.scatter(x,y)
plt.savefig("scatter.svg")
plt.show()
```

```python
## hisstogram
age=[18,22,55,62,32]
b=[0,10,20,30,40]
plt.hist(age,b,histtype='bar',rwidth=0.5)
plt.show()
```

```python
from numpy.random import normal
x=normal(size=200)
plt.hist(x,bins=30)
plt.show()
```
